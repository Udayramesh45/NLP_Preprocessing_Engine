```python
## Task 2:Build Advanced Preprocessing Function

import re
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Remove emojis & special characters
    text = re.sub(r"[^\w\s]", "", text)

    # Handle repeated characters (soooo → soo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    tokens = text.split()

    # Remove short tokens (keep 'no', 'not')
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    cleaned_sentence = " ".join(cleaned_tokens)


    return cleaned_tokens, cleaned_sentence
```

```python
## Task 3: Stress Testing

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)
    results.append((sentence, tokens, clean_sentence))

for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Cleaned:", r[2])
```

```python
## Task 4: Token Analytics

def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)


for sentence, tokens, clean_sentence in results:
    total, unique, avg = token_analytics(tokens)

    print("\nSentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", avg)
```

```python
##Task 5: Frequency Analysis
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

```python
## Task 6: Build Full Pipeline
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


output = full_pipeline(test_sentences)

print(output)
```

```python
##Task 7: Error Handling
edge_cases = [
    "",
    "😍😍😍😍",
    "1234567890"
]

for text in edge_cases:
    tokens, clean = preprocess_text(text)
    print("\nInput:", text)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
```
